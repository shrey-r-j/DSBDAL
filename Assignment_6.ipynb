```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
df.dtypes
```

```python
nfeat = ["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"] 
cfeat = "Species"
```

```python
plt.figure(figsize=(10,6))

for i in nfeat:
    plt.title(f"Histogram of the feature :-{i}")
    sns.histplot(df[i],bins=10,kde=True)
    plt.show()


```

```python
plt.figure(figsize=(10,6))

for i in nfeat:
    plt.title(f"Boxplot of the feature :-{i}")
    sns.boxplot(df[i])
    plt.show()
```

```python
plt.figure(figsize=(10,8))

for i in nfeat:
    plt.title("Swarm Plot of Petal Length")
    # sns.countplot(x=df[cfeat],hue=df[i])
    sns.swarmplot(x=cfeat, y=i, data=df)
    plt.show()
```

```python
plt.figure(figsize=(10,8))

for i in nfeat:
    plt.title("Violin Plot of Petal Length")
    # sns.countplot(x=df[cfeat],hue=df[i])
    sns.violinplot(x=cfeat, y=i, data=df)
    plt.show()
```

```python
plt.figure(figsize=(10,8))

for i in nfeat:
    plt.title("KDE Plot of Petal Length")
    # sns.countplot(x=df[cfeat],hue=df[i])
    sns.kdeplot(df[i])
    plt.show()
```
