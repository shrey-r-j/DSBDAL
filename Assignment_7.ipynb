```python
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords,wordnet
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import string
import math
from collections import Counter
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
document = """Natural Language Processing (NLP) is a field of Artificial Intelligence
that enables computers to understand human language. NLP is widely used
in chatbots, translation systems, and text analytics applications."""
print(document)
```

```python
tokens = word_tokenize(document.lower())
print(tokens)
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [
    word for word in tokens
    if word.isalpha() and word not in stop_words
]

print(filtered_tokens)
```

```python
pos_tags = pos_tag(filtered_tokens)
print("POS Tags:\n", pos_tags)
```

```python
stemmer = PorterStemmer()
stemmed_words = [
    stemmer.stem(word) for word in filtered_tokens
]
print(stemmed_words)
```

```python
#Lemmetizer

def get_wordnet_pos(tag) :
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('R'):
        return wordnet.ADV
    elif tag.startswith('N'):
        return wordnet.NOUN
    else : return wordnet.NOUN

lemmatizer = WordNetLemmatizer()
lemmatized_words = [
    lemmatizer.lemmatize(word,get_wordnet_pos(tag))
    for word,tag in pos_tags
]
print(lemmatized_words)
```

```python
tf = Counter(filtered_tokens)

for word , freq in tf.items():
    print(word,":",freq)
```

```python
documents = [document]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
print("Feature names\n",vectorizer.get_feature_names_out())
print("\nFeature values\n",tfidf_matrix.toarray())
```
